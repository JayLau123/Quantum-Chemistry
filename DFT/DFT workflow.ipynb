**Schordinger**

matter wave $\Rightarrow$ wave equation: Schordinger's equation(S.eq)

$$\left[-\dfrac{\hbar^2}{2 m} \nabla^2+V(\vec{r})\right] \psi(\vec{r}, t)=i \hbar \dfrac{\partial}{\partial t} \psi(\vec{r}, t)$$

**How to solve S.eq(non-relativistic, time-independent)**

$$\hat{H}\psi = \hat E \psi$$

Each solution $\psi_n$ has an associated eigenvalue $E_n$ it's a real number because $\hat H$ is a Hermite operator. $\psi_n$ and $E_n$ satisfies the eigenvalue equation.

$\hat H=\hat T+\hat V$: contains all the information about the kinetic and potential energier for all particles of a many-body system.

Some solvable examples with analytic solutions:
1. Hydrogen atom
2. 1D harmonic oscillator
3. 1D infinite potential
4. Particle in magnetic field

However, in reality, what we interested are various of molecules, materials that are much more complicated.





Hamiltonian operator of a many-body system, without external electronmagnetic field, has 5 terms. Theoretically, it's correct but inefficient to solve S.eq, how to simplify the $\hat H$ and solve S.eq in a smart way?

**BornOppenheimer approximation**

Premise: The mass difference of electron and nuclei. Given the same kinetic energy, from the perspective of an electron, nuclei positions are fixed approximately.

Therfore, divide the system hamiltonian $\hat H$ into two parts that can be solved seperately. We can solve the electronic part repeatedly while the nuclei positions are changed in a small step. The electronic energy can be regarded as a function of the nuclei coordinates to produce a potential energy surface. 

As a result, the kinetic energy of nuclei is zero, and their potential is a constant. 

$$\hat{H}=-\sum_{i=1}^N \dfrac{\hbar^2}{2 M e} \nabla_i^2-\sum_{A=1}^M \dfrac{\hbar^2}{2 M_A} \nabla_A^2+\sum_{i=1}^N \sum_{j=1+1}^N k \dfrac{e^2}{r_{i j}}-\sum_{i=1}^N \sum_{A=1}^M k \dfrac{z_A e^2}{r_{i A}}+\sum_{i=1}^M \sum_{E_A=+1}^M k \dfrac{z_0 Z_\theta e^2}{R_{A B}}$$

$\Rightarrow$

$$\hat{H}=-\sum_{i=1}^N \dfrac{\hbar^2}{2 M e} \nabla_i^2-0+\sum_{i=1}^N \sum_{j=1+1}^N k \dfrac{e^2}{r_{i j}}-\sum_{i=1}^N \sum_{A=1}^M k \dfrac{z_A e^2}{r_{i A}}+constant$$

$$\approx -\sum_{i=1}^N \dfrac{\hbar^2}{2 M e} \nabla_i^2+\sum_{i=1}^N \sum_{j=1+1}^N k \dfrac{e^2}{r_{i j}}-\sum_{i=1}^N \sum_{A=1}^M k \dfrac{z_A e^2}{r_{i A}}$$

$$=T_e+V_{ee}+V_{eN}$$

= kinetic energy of electrons+electrostatic attraction+electrostatic replusion

In summary, the molecular system can be described by electronic hamiltonian $\hat H$, and we can solve the simplified S.eq to obtain electronic wave function $\psi_e$, which only depends on electronic spatial coordinates(spins are neglected).

P.S the $\psi$ that gives the lowest $E$ is ground state, and this $E$ is ground state energy.

However, it's still difficult to solve $psi$, because there are many electron-electron terms and it's a many-body problem, which means all electrons must be considered simultaneously.

Actually, the $\psi$ can't be directly measured in principle. What can be measured is the probability that there are $N$ electrons at certain coordinates $\vec{r_1}, \vec{r_2}, \vec{r_3}...$

The probability density is given by $|\psi|^2=\psi^*(\vec{r_1}, \vec{r_2}, \vec{r_3}...)\psi(\vec{r_1}, \vec{r_2}, \vec{r_3}...)$, where * means complex conjugate.

**Hartree product**

$\psi(\vec{r_1}, \vec{r_2}, \vec{r_3}...)$ can be approximately written as a product of individual electron wave functions:

$\psi(\vec{r_1}, \vec{r_2}, \vec{r_3}...)\approx \psi(\vec{r_1})\psi (\vec{r_2})...$

However, electrons are indistinguishable from experiments, so we can't really label them like electron 1, electron 2, electron N...

** Electron density**

Density of electrons at a particular position in space.

Electron density can be calculated by counting how many electrons will we find at this position **on average**.

$\rho(\vec{r})=2\sum_i \psi_i^*(\vec{r})\psi_i(\vec{r})$, where '2' means electrons have spin up or down. Pauli exclusion principle shows each electronic wave function can be occupied by two seperate electrons with opposite spins. So we can find the electron density by adding all the probabilities.

Before: $\psi(\vec{r_1}, \vec{r_2}, \vec{r_3}...) is a function of 3N variables$
After: $\rho(\vec{r})$ is a function of 3 coordinates(no spins)

Multi-electron wave function: $\psi(\vec{r_1}, \vec{r_2}, \vec{r_3}...\vec{r_n}) $
 

$\psi^*\psi=|\psi|^2$=probability density that at 't' moment, 1st particle is in $\vec{r_1}$,  2nd particle is in $\vec{r_2}$,  3rd particle is in $\vec{r_3}...$

Probability=$P=\int |\psi|^2dr=\int...\int|\psi|^2dr_1dr_2dr_3...dr_n$ **(n multiple integral)**

The probability density that one electron shows in $\vec{r_1}$: $\int...\int|\psi|^2dr_2dr_3dr_4...dr_n$ **(n-1 multiple integral)**

Electrons are indistinguishable, considering total $n$ electrons and $2$ spins, the electron density at $\vec {r_1}$:

$\rho(\vec{r_1})=2N \int |\psi|^2dr_2dr_3...dr_n$ **(n-1 multiple integral)**, where $\psi=\psi(\vec{r_1}, \vec{r_2}, \vec{r_3}...\vec{r_n})$

$\rho(\vec{r})$: a function of only 3 coordinates but contains information that is actually observable from the full wave function. So we can replace wave function with electron density.

**Theory background: Kohn, Hohenberg, Sham**

The ststem energy can be represented as the functional of electron density.

1. The ground-state energy $E$ is a unique functional of electron density $\rho(\vec r)$

One-to-one mapping between the ground state wave function and the ground state electron density, $E[\rho(\vec r)]=E$, $E$ can be expresed as a functional of electron density.

2. The electron density that minimizes the energy of the overall functional is the true electron density, corresponding to the full solution of the S.eq. Instead of solving the S.eq to find a function of $3N$ variables, we can just define a function of just 3 variables

The complete energy functional

$E[\psi_i]=E_{known}[\psi_i]+E_{xc}[\psi]$, where the first term is single electron wave function, and the third term is exchange-correlation term. 

The second term: $E_{known}[\psi_i]$ (4 contributions): electron kinetic eneergy, electrostatic interactions between electrons and nuclei, electrostatic repulsion between pair of electrons, electrostatic repulsion between pair of nuclei.

$E_{xc}$ exchange-correlation term includes all quantum mechanical effects beyond known part.

3. ALl properties of the system are functionals of the ground state density $\rho_0(\vec r)$

4. For any system of the interacting particles, the $E_{xc}$ is unique.

**How to find the minimum energy?**

**Kohn-Sham equation**: find the right electron density by solving a set of equations, each involving only a single electron

$$[\dfrac{1}{2}\nabla^2+V(\vec r)+V_H(\vec r)+V_{xc}(\vec r)]\psi_i(\vec r)=\epsilon_i \psi_i(\vec r)$$

P.S. the left part of the equation is a single electronic hamiltonian. Atomic unit, only one electron, no summation

$V(\vec r)$: potential term. Interaction between an electron and a collection of atomic nuclei.

$V_H(\vec r)$: Hartree potential. $V_H(\vec r)=\int \dfrac{\rho(\vec r')}{|\vec r-\vec r'|}d^3r' $

Describe the coulomb repusion between the electron being considered in one of Kohn-Sham orbitals,  and the total electron density defined by all electrons in the system. Hartree potential contains a self-interaction contribution.


$V_{xc}(\vec r)$: potential that defines the exchange and correlation contributions to the single-electron equations.

$V_{xc}(\vec r)=\dfrac{\delta E_{xc}(\vec r)}{\delta \rho(\vec r)}$. $V_{xc}(\vec r) $can be defined as a functional derivative of the exchange correlation energy with respect to the electron density.


**How to solve Kohn-Sham equation**

1. To solve the K-S equations, requires Hartree potential $V_H(\vec r)=\int \dfrac{\rho(\vec r')}{|\vec r-\vec r'|}d^3r' $
2. To find the Hartree potential, requires electron density $\rho(\vec r)=2 \sum_i \psi_i^*(\vec r)\psi_i(\vec r)$
3. To find electron density, requires single electron wave function $\psi_i(\vec r)$
4. To find single electron wave function, requires solving K-S equations.

**We can use iterative method and initial conditions to break the circle**

1. Define an initial electron density $\rho_0(\vec r)$
2. Use $\rho_0(\vec r)$ to solve K-S equation and find the single-electron wave function $\psi_i(\vec r)$
3. Calculate electron density $\rho(\vec r)=2 \sum_i \psi_i^*(\vec r)\psi_i(\vec)$
4. Compare the calculated $\rho(\vec r)$ with previous one $\rho_0(\vec r)$, if $|\rho(\vec r)-\rho_0(\vec r)|<\epsilon$. They are the same approximately, that is the ground-state electron density. Otherwise update the initial $\rho_0(\vec r)$ in some way and repeat again.

Once we find true $\rho(\vec r)$, we obtain information that's actually physically observable from the wave function.


**How to specify the exchange correlation functional**

However, the true form is unknown, and must be approximated. There is one case where XC functional can be exactly derived: Uniform electron gas. In uniform electron gas, electron density is a constant everywhere, but it's not realistic in other situations.


The most common approximations of XC functional: **LDA, GGA, hybrid functional**

**LDA**: local density approximation

$V_{xc}(\vec r)=V_{xc}[\rho(\vec r)]$

We can set the exchange correlation potential at each position to be the known XC potential from a uniform electron gas at the electron density observed at that position. LDA uses the local density to define the approximate XC functional.

**GGA**: generalized gradient approximation

GGA uses information about local electron density and the local gradient in the electron density. The most commonly used GGA functionals for solids are **PW91(Perdew-Wang 91)**, or **PBE(Perdew-Burke-Ernzerhof)** functional. For solid, different crystall structure and different configuration of atoms requires different functional.




P.S. Quantum chemistry calculations can be categorized by the type of functions: Spatially localized(cluster-type calculations) and Spatially extended(bulk-type calculations). Methods based on spatially extended, periodic functions(VASP calculations) are known as plane-wave methods
